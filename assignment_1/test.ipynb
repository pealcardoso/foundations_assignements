```python
import pandas as pd

df=pd.read_csv('../life_expectancy/data/eu_life_expectancy_raw.tsv', sep='\t|,')
df
```

```python
df2=pd.melt(df, id_vars=df.columns[0:4], value_vars=df.columns[4:], var_name='date')
df2.rename(columns={'geo\\time':'region'},inplace=True)
df2 = df2.astype({"date": int})
#df2 = df2[(df2.value!=':') & (df2.value!=': ')]
df3=df2.copy()
df2.value = df2.value.str.extract(r"(\d+\.\d+)")
df2.dropna(inplace=True)
df2 = df2.astype({"value": float})
df2.dtypes
```

```python
portugal=df2[df2.region=='PT']
portugal.to_csv('../life_expectancy/data/pt_life_expectancy.csv',index=False)
```
